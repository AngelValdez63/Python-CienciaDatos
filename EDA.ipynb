los datos son Longitud-sepalo, anchura-sepalo, longitud-petalo y anchura-sepalo.

```python
##Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("bmh")                    #Estilo predeterminado de los graficos
sns.set(style="whitegrid")              #Fondo con cuadricula blanca

import pandas as pd
df = pd.read_csv("/content/IRIS - IRIS.csv")       #Cargamos el data set y lo imprimimos
df
```

Preprocesamiento
Renombramos los nombres de las columnas para mas facilidad de codigo.

```python
df.rename({'sepal-length': 'sep_l',
           'sepal-width':'sep_a',
           'petal-length':'pet_l',
           'petal-width':'pet_a',
           'species':'Especies'},
          axis=1,inplace=True)
df
```

Analisis univariado

Longitud del sepalo

```python
#histograma con kde
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.histplot(df['pet_l'],kde=True,bins=20,color='purple')
plt.title('Distribucion de la longitud del sepalo')
plt.xlabel('Longitud del sepalo')
plt.ylabel('Frecuencia')

#boxplot
plt.subplot(1,3,2)
sns.boxplot(y=df['pet_l'],color='lightgreen')
plt.title('Boxplot de la longitud del sepalo')
plt.xlabel('Longitud del sepalo en cm')
plt.ylabel('')

#violinplot
plt.subplot(1,3,3)
sns.violinplot(y=df['pet_l'],color='lightblue')
plt.title('Violinplot de la longitud del sepalo')
plt.xlabel('Longitud del sepalo en cm')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
#histograma con kde
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.histplot(df['pet_a'],kde=True,bins=20,color='purple')
plt.title('Distribucion del ancho del sepalo')
plt.xlabel('Longitud del sepalo')
plt.ylabel('Frecuencia')

#boxplot
plt.subplot(1,3,2)
sns.boxplot(y=df['pet_a'],color='lightgreen')
plt.title('Boxplot del ancho del sepalo')
plt.xlabel('Longitud del sepalo en cm')
plt.ylabel('')

#violinplot
plt.subplot(1,3,3)
sns.violinplot(y=df['pet_a'],color='lightblue')
plt.title('Violinplot del ancho del sepalo')
plt.xlabel('Longitud del sepalo en cm')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

Analisis multivariable

```python
fig, axs = plt.subplots(2,2, figsize=(12,14))


# Histograma longitud del sepalo por especie
sns.histplot(data=df, x='sep_l',hue='Especies', multiple='stack', palette='Set1',bins = 20, ax=axs[0,0])
axs[0,0].set_title('Distribucion de la longitud del sepalo por especie')
axs[0,0].set_xlabel('Longitud del sepalo')
axs[0,0].set_ylabel('Frecuencia')


#Diagrama de caja (Boxplot) de longitud del sepalo por especie
sns.boxplot(data=df, x='Especies',y='sep_l', palette='Set1',ax=axs[0,1])
axs[0,1].set_title('Boxplot de la longitud del sepalo por especie')
axs[0,1].set_xlabel('Especie')
axs[0,1].set_ylabel('Longitud del sepalo')

#Grafico de volin de longitud del sepalo  por especie
sns.violinplot(data=df, x='Especies', y='sep_l', palette='Set1',ax=axs[1,0])
axs[1,0].set_title('Violinplot de la longitud del sepalo por especie')
axs[1,0].set_xlabel('Especie')
axs[1,0].set_ylabel('Longitud del sepalo')

#Grafico de dispersion (Scatter plot) de longitud del sepalo por especie
sns.scatterplot(data=df, x='sep_l', y='pet_l', hue='Especies', palette='Set1',ax=axs[1,1])
axs[1,1].set_title('Grafico de dispersion de longitud de sepalo y petalo por especie')
axs[1,1].set_xlabel('Longitud del sepalo')
axs[1,1].set_ylabel('Longitud del petalo')

plt.tight_layout()
plt.show()
```

Relacion entre variables

```python
sns.pairplot(df,hue='Especies',markers=["o","s","d"],palette='Set1')
plt.suptitle('Pairplot dataset IRIS', y=1.02)
plt.show()
```

Correlacion de variables

```python
plt.figure(figsize=(8,6))
correlation_matrix = df.drop(columns=['Especies']).corr()
sns.heatmap(correlation_matrix,annot=True, cmap='coolwarm',vmin=-1,vmax=1)
plt.title('Matriz de correlacion')
plt.show()
```

**Conclusiones.**

Con los datos recolectados se llega a la conlcusion que el ancho y la longitud de los petalos pueden utilizarse para saber la especie de la flor. lo podemos observar en el histograma, en los boxplot una significativa diferencia entre los tamaños de las especies, tambien vemos correlaciones entre las variables. observamos tanto en el heatmap como en el pairplot la relacion entre estos.
